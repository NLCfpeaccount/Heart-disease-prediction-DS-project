```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import * 
```

```python
d=pd.read_csv('heart-disease.csv')

'''
d=pd.read_csv('11zon_zip/heart-disease.csv')
 with pd.option_context('display.max_rows', None,
                           'display.max_columns', None,
                           'display.precision', 3,
                           ):print(d)
'''
d.isna().cumsum()
```

```python
                          

#features and matrix
x=d.drop('target',axis=1)
y=d['target']
```

```python
import sklearn
xtr,xtst,ytr,ytst= sklearn.model_selection.train_test_split(x,y,test_size=0.2)
```

```python
clf=sklearn.ensemble.RandomForestClassifier(n_estimators=23)


clf.fit(xtr,ytr)
 #model validate
clf.score(xtr,ytr)
 
```

```python
print(clf.get_params())
result=clf.predict(xtst)
result 
```

```python
import sklearn.metrics as est 
```

```python
print(est.classification_report(ytst,result))
```
